```python
'''
1) Write a Python program to check whether a given word is a palindrome.

Example Input:
"radar"

Example Output:
True
'''

palindrome = input("Enter a word: ")
if palindrome == palindrome[::-1]:
    print("It's a palindrome.")
else:
    print("It's not a palindrome.")
```

```python
'''
4) Write a Python program to demonstrate multi-level inheritance. Create three classes: Grandfather, Father, and Rohan (the child class).
The Grandfather class should have some properties, and these properties should be passed down to the Father class.
The Father class will pass those properties to the Rohan class. Finally, call the properties from both the Child (Rohan) and Grandfather classes inside the Rohan class.
'''

class Grandfather:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show(self):
        print(f"Grandfather's Name: {self.name}, Age: {self.age}")

class Father(Grandfather):
    def __init__(self, name, age, address):
        super().__init__(name, age)
        self.address = address

    def show(self):
        super().show()
        print(f"Father's Address: {self.address}")

class Rohan(Father):
    def __init__(self, name, age, address, grade):
        super().__init__(name, age, address)
        self.grade = grade

    def show(self):
        super().show()
        print(f"Rohan's Grade: {self.grade}")

rohan = Rohan('Aaditya', 70, 'Indore', 78)
rohan.show()
```

```python
'''
5) Write a Python program that asks the user for a number and then checks whether the number is even, odd, or prime.
Finally, print the result indicating whether the number is even, odd, or prime.
'''

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
no = int(input("Enter a number: "))
if no % 2 == 0:
    print(f"{no} is even")
elif is_prime(no):
    print(f"{no} is prime")
else:
    print(f"{no} is odd")
```

```python
'''
6) Write a Python program to print a diamond pattern using stars (*). The program should ask the user to input the number of rows for the top half of the diamond and then print the pattern accordingly.

    *
   ***
  *****
 *******
*********
 *******
  *****
   ***
    *
'''

for i in range(1, 10, 2):
    print(" " * ((9 - i) // 2) + "*" * i)

for i in range(7, 0, -2):
    print(" " * ((9 - i) // 2) + "*" * i)
```

```python
'''
7) Write a Python program that demonstrates data visualization using the matplotlib library. The program should:

Create a simple dataset (e.g., sales data for a week or the population of different countries).
Plot this data using different types of visualizations:
A bar chart to compare values.
A line chart to show the trend over time.
A pie chart to represent proportions.
'''

import matplotlib.pyplot as plt

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sales = [120,340,235,675,753,275,542]

plt.figure(figsize=(10,5))
plt.bar(days, sales)
plt.title("Sales Data")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(days, sales)
plt.title("Sales Data")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(8, 8))  # Set the figure size for the pie chart
plt.pie(sales, labels=days, autopct="%1.0f%%")
plt.title("Pie chart for sales")
plt.show()
```
